### PROBLEME DE N REINES

    On considère un échiquier carré de taille N x N. On dispose de N reines. Le problème consiste à placer les N reines de façon à ce que deux reines quelconques ne s’attaquent pas mutuellement. Rappelons qu’une reine attaque sur les lignes, colonnes et les diagonales.
On utilisera un tableau dynamique à deux dimensions d’entiers. On utilisera les valeurs suivantes :

• 0 : pour indiquer une case libre.

• 1 : pour indiquer une case interdite.

• y : pour indiquer une case occupée par la reine numéro y (y = 1..n)

### SOLUTION EN PYTHON

<h5 style="color:blue;"> 1. Variables globales: </h5>

```python
echiquier=[] #Tableau de l'echiquier de taille NxN
position_Reine=[] #Talbeau de position des N reines
```

<h5 style="color:blue;"> Precedure d'initialisation : </h5>

>Initialiser le tableau de l'echiquier de taille N*N par des 0

>Initialiser le tableau contenant les positions des N reines par des 0

```python
def initialiser_Echiquier():
    for i in range(N):
        position_Reine.append(0)
        li=[]
        for k in range(0,N):
            li.append(0)
        echiquier.append(li)
```

<h5 style="color:blue;"> Afficher() : </h5>

>procédure permet d’afficher l’echiquier 

```python
def affiche():
    for element in echiquier :
        print (element)
    print ("__________")     
```

<h5 style="color:blue;">Poser() : </h5> 

>procédure permit de remplir toutes les cases vides de l'échiquier par 0, et les cases contenant les reines par le numéro de raine et on affiche le résultat.

```python
def poser():
    for i in range(N):
        for k in range(0,N):
            echiquier[i][k]=1
    for i in range(0,N): 
        echiquier[i][position_Reine[i]]=i+1
    affiche()
```

<h5 style="color:blue;">EstLibre() : </h5>

>fonction permet de tester si la case à la ligne ligne et à la colonne colonne n'est pas menacée par les reines précédemment placées.

```python
def estLibre(l ,c): 
    for i in range(0,c):
        if position_Reine[i]==l or abs(position_Reine[i]-l)==abs(i-c):
            return 0 
    return 1 
```

<h5 style="color:blue;">placerReine() :</h5>

```python
def placerReine(cur_col): 
    if(cur_col==N): 
        poser()      
    else :
        '''pour une colonne donne, on teste si l'une des ligne est valide pour y poser la reine,
           si aucune ligne n'est valide, on retourne a la colonne precedante et change la position de la reine qui s'y trouve
        '''
        for i in range(0,N): 
                if(estLibre(i,cur_col)): #teste si la case est valide
                    position_Reine[cur_col]=i 
                    placerReine(cur_col+1) #on passe a la colonne suivante
```
