   # Générer un tableau de n entier aléatoire 

```python
import random
import numpy as np
def createVect(N):
    vect = np.random.randint(50, size=(N))
    return vect 
```

```python
import matplotlib.pyplot as plt
from datetime import datetime
N=[10, 500, 1000,1500, 2000, 5000, 10000]
```

  
# 1. Tri par selection

```python
def triSelection(v):
    for i in range(len(v)):
        ind_min=i
        for j in range(i+1, len(v)):
            if(v[ind_min]>v[j]):
                ind_min=j
        aux=v[i]
        v[i]=v[ind_min]
        v[ind_min]=aux
    return v
```

```python
def calcule_Tri_Selection(N) :
    time=[]
    for i in N:
        vect = createVect(i)
        start_time = datetime.now()
        triSelection(vect)
        end_time = datetime.now()
        time.append(format(end_time - start_time))
        
    t=time
    print("les mésures d'execution de tri par selection : ")

    plt.plot(N,time,'o-', color="red", linewidth=2.5, linestyle='-')
    plt.grid(True)
    plt.ylabel('temps')
    plt.xlabel('Taille')
    plt.title('Mesure d''execution TRI Selection')
    plt.show
```

```python
calcule_Tri_Selection(N)
```

# 2. Tri a bulles

```python
def tri_Bulle(t):
    n = len(t)
    for i in range(n):
        for j in range(0, n-i-1):
            if t[j] > t[j+1] :
                aux=t[j]
                t[j]=t[j+1]
                t[j+1]=aux
```

```python
def calcule_Tri_Bulle(N) :
    time=[]
    for i in N:
        vect = createVect(i)
        start_time = datetime.now()
        tri_Bulle(vect)
        end_time = datetime.now()
        time.append(format(end_time - start_time))
        
    t=time
    

    plt.plot(N,time,'o-', color="red", linewidth=2.5, linestyle='-')
    plt.grid(True)
    plt.ylabel('temps')
    plt.xlabel('Taille')
    plt.title('Mesure d''execution TRI Bulle')
    plt.show
```

```python
calcule_Tri_Bulle(N)
```

# 3. Tri par Insertion

```python
def triInsertion(t): 
    for i in range(1, len(t)): 
        k = t[i] 
        j = i-1
        while j >= 0 and k < t[j] : 
                t[j+1] = t[j] 
                j -= 1
        t[j+1] = k
```

```python
def calcule_Tri_Insertion(N) :
    time=[]
    for i in N:
        vect = createVect(i)
        start_time = datetime.now()
        triInsertion(vect)
        end_time = datetime.now()
        time.append(format(end_time - start_time))
        
    t=time
    print("les mésures d'execution de tri par Insertion : ")

    plt.plot(N,time,'o-', color="red", linewidth=2.5, linestyle='-')
    plt.grid(True)
    plt.ylabel('temps')
    plt.xlabel('Taille')
    plt.title('Mesure d''execution TRI Insertion')
    plt.show
```

```python
calcule_Tri_Insertion(N)
```

# 4.  Tri fusion

```python
def triFusion(L):
    n = len(L)
    if n > 1:
        p = int(n/2)
        L1 = L[0:p]
        L2 = L[p:n]
        triFusion(L1)
        triFusion(L2)
        L[:] = fusion(L1,L2)
```

```python
def fusion(L1,L2):
    n1 = len(L1)
    n2 = len(L2)
    L12 = [0]*(n1+n2)
    i1 = 0
    i2 = 0
    i = 0
    while i1<n1 and i2<n2:
        if L1[i1] < L2[i2]:
            L12[i] = L1[i1]
            i1 += 1
        else:
            L12[i] = L2[i2]
            i2 += 1
        i += 1
    while i1<n1:
        L12[i] = L1[i1]
        i1 += 1
        i += 1
    while i2<n2:
        L12[i] = L2[i2]
        i2 += 1
        i += 1 
    return L12
```

```python
def calcule_Tri_Fusion(N) :
    time=[]
    for i in N:
        vect = createVect(i)
        start_time = datetime.now()
        triFusion(vect)
        end_time = datetime.now()
        time.append(format(end_time - start_time))
        
    t=time
    

    plt.plot(N,time,'o-', color="red", linewidth=2.5, linestyle='-')
    plt.grid(True)
    plt.ylabel('temps')
    plt.xlabel('Taille')
    plt.title('Mesure d''execution TRI Fusion')
    plt.show
```

```python
calcule_Tri_Fusion(N)
```

# 5.  Tri rapide

```python
def triRapide(t):
    if t == []:
        return []
    else:
        pivot = t[0]
        t1 = []
        t2 = []
    for x in t[1:]:
        if x<pivot:
            t1.append(x)
        else:
            t2.append(x)
    return triRapide(t1)+[pivot]+triRapide(t2)
```
